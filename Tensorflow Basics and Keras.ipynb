```python
import pandas as pd
```

```python
import numpy as np
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("./fake_reg.csv")
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df[['feature1','feature2']].values
```

```python
X = df[['feature1','feature2']].values
```

```python
y = df['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=11)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_train
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))

model.add(Dense(1))
model.compile(optimizer='rmsprop',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,epochs=500)
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

```python
model.evaluate(X_test,y_test)
```

```python
model.evaluate(X_train,y_train)
```

```python
test_predictions = model.predict(X_test)
test_predictions = pd.Series(test_predictions.reshape(200,))
```

```python
pred_df =pd.DataFrame(y_test,columns=['Test True Y'])
```

```python
pred_df = pd.concat([pred_df,test_predictions],axis=1)
```

```python
pred_df.columns = ['Test True Y','Model Predictions']
```

```python
pred_df
```

```python
sns.scatterplot(x="Test True Y",y="Model Predictions",data=pred_df)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
print(mean_absolute_error(pred_df['Test True Y'],pred_df['Model Predictions']))
```

```python
print(mean_squared_error(pred_df['Test True Y'],pred_df['Model Predictions']))
```
