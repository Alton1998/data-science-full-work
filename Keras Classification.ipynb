```python
X=df.drop('benign_0__mal_1',axis=1).values
```

```python
y=df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense,Dropout
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),batch_size=32)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
model1 = Sequential()
model1.add(Dense(30,activation='relu'))
model1.add(Dense(15,activation='relu'))

model1.add(Dense(1,activation='sigmoid'))
model1.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
help(EarlyStopping)
```

```python
early_stop = EarlyStopping(monitor='val_loss',mode="min",verbose=1,patience=25)
```

```python
model1.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),batch_size=32,callbacks=[early_stop])
```

```python
losses = pd.DataFrame(model1.history.history)
```

```python
losses.plot()
```

```python
model2 = Sequential()
model2.add(Dense(30,activation='relu'))
model2.add(Dropout(0.5))
model2.add(Dense(15,activation='relu'))
model2.add(Dropout(0.5))
model2.add(Dense(1,activation='sigmoid'))
model2.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model2.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),batch_size=32,callbacks=[early_stop])
```

```python
losses = pd.DataFrame(model2.history.history)
```

```python
losses.plot()
```

```python
preds=model2.predict(X_test)
classes_x=(preds > 0.5).astype(np.int8)
```

```python
classes_x
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,classes_x))
```

```python
print(confusion_matrix(y_test,classes_x))
```
