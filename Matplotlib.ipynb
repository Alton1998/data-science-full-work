```python
import matplotlib.pyplot as plt
```

```python
# Lets you see diagrams in jupyter
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional way of using matplotlib
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y label')
# plt.show() #doesn't need to be used in this case only when you are not using jupyter you should uses this
```

```python
plt.subplot(1,2,1) # row, column and index
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented Method
fig = plt.figure()

# Adding axes
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # [left, bottom, width, height]
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title("Sample graph")
```

```python
fig1 = plt.figure()
axes1 = fig1.add_axes([0.1,0.1,0.8,0.8])# created to an empty canvas
axes2 = fig1.add_axes([0.2,0.2,0.6,0.3])# created relative to the previous canvas

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
# Creating subplots

fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
# Figure Size and DPI
fig2 = plt.figure(figsize = (8,2))

ax2=fig2.add_axes([0,0,1,1])
ax2.plot(x,y)
```

```python
fig3,axes3 = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes3[0].plot(x,y)
axes3[1].plot(y,x)
plt.tight_layout()
```

```python
fig3.savefig('my_picture.png',dpi=200)# Saving a plot
```

```python
fig4 = plt.figure()

ax4 = fig4.add_axes([0,0,1,1])
ax4.plot(x,x**2,label='X Squared',color='#FF8CC0',linewidth=10,alpha=0.5,linestyle='--',marker='o',markersize=20,markerfacecolor='red',markeredgecolor='green')# RGB Hex Code
ax4.plot(x,x**3,label ='X cubed')
ax4.legend(loc=0)
ax4.set_xlim([0,2]) ## setting bounds on the x axis
ax4.set_ylim([0,20])
```
