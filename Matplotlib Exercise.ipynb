```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(1,10,10)
y = np.linspace(1,10,10)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title("Title")
ax.set_ylabel("Y axis")
ax.set_xlabel("X axis")
ax.plot(x,y)
```

```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
ax2 = fig1.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x,y)
ax2.plot(x,y)
```
